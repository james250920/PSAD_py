```python
import yt_dlp

def descargar_video_yt_dlp(url):
    ydl_opts = {}
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(url, download=False)
        print(f"Titulo: {info_dict['title']}")
        print(f"Descripcion: {info_dict['description']}")
        print(f"Duracion: {info_dict['duration']} segundos")
        print(f"Numero de vistas: {info_dict['view_count']}")
        
        opcion = input("Desea descargar el video en alta calidad? (s/n): ")
        if opcion.lower() == "s":
            ydl_opts = {'format': 'best'}
        else:
            ydl_opts = {'format': 'worst'}
        
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        print("Descarga completa")

url = input("Ingrese la URL del video de YouTube que desea descargar: ")
descargar_video_yt_dlp(url)
```
